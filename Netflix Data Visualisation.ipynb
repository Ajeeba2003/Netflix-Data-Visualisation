# Introduction

The netflix.csv dataset offers a detailed view of Netflix's extensive content library, including movies, TV shows, and other media. With attributes like titles, release years, genres, ratings, durations, and countries of origin, this dataset provides a rich foundation for exploring Netflix's content trends and patterns

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('netflix.csv')
data
```

# About Data
> # Column Description
  * show_id:Unique id for each shows
  * type:Type of the show.(Movie,TV Show)
  * title:Name of each show.
  * director:Name of the director of the show.
  * cast:Name of the cat members or actors of particular show.
  * country:Name of the country released the show.
  * date_added:date of data added to the dataframe.
  * release_year:Date of release of the show
  * rating:Rating of the show.
  * duration:Duration of the shows
  * listed_in:The genere the movie belongs to.
  * description:A small description about the show

```python
data['date_added']=data['date_added'].astype('datetime64[ns]')
```

```python
data.info()
```

```python
data['genre']=data['listed_in'].str.split(',').str[0]
data
```

```python
data.assign(cast=data['cast'].str.split(',')).explode('cast')
```

```python
# 1.distibution of movie and TV shows:
data['type'].value_counts().plot(kind='bar',color='brown')
plt.title('Distribution of Movies and TV Show')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

* Most of the show released by netflix is movies
* As compared to movies ,TV shows are rarely raleased in netflix

```python
#2. number of movies released in each year:
a=data[data['type']=='Movie']
a.groupby(['release_year'])['title'].count().plot(kind='line',marker='s',color='green')
plt.title('Trend of movie released in each year')
plt.show()
```

* There is a sudden increase in the number of movies released from 2010 to 2020 and is dropped in 2020

```python
#3.top 10 genre of movie released in netflix:
a['genre'].value_counts().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.2f%%')
plt.title('Distribution of Genre under Movies')
plt.show()
```

* most of genre of the movies comes under drama

```python
#4.Distribution of rating:
data['rating'].value_counts().plot(kind='bar',color='yellow',edgecolor='black')
plt.title('Distribution of rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

* Most of the content of netflix comes under TV-MA and TV-14

```python
#5. Distribution of released year:
sns.histplot(x='release_year',data=data)
plt.title('Distribution of released year')
plt.show()
```

* A lot of contents were released in 2010 to 2020

```python
# 6.top ten countries contributing to netflix:
b=data.assign(country=data['country'].str.split(',')).explode('country')
b['country']=b['country'].str.lstrip(' ')
b['country'].value_counts().sort_values(ascending=False).head(20).plot(kind='bar',color='pink',edgecolor='red')
plt.title('Top 10 Countries Contributing to Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

* The graph top 10 countries contributing content to Netflix 
* United states contributed more contents to Netflix
* India takes second place in contributing to Netflix
* Least contibution is from italy

```python
# 7.distibution of top 10 directors
data['director'].value_counts().sort_values(ascending=False).head(20).plot(kind='bar',color='orange') 
plt.title('Top 10 Contributed to Netflix ')
plt.show()
```

* the graph shows top 10 directors contributed to netflix
* Raul Campos,Jan suter directed most of the show in netflix


```python
#8. Duration of movie
a=data[data['type']=='Movie']
```

```python
def split(x):
    return x.split(' ')[0]
```

```python
a['duration']=a['duration'].apply(split)
a['duration']=a['duration'].astype('int64')
```

```python
a.groupby(['title'])['duration'].max().sort_values(ascending=False).head(10).plot(kind='bar',color='red')
plt.title('Duration of movie in minutes')
plt.xlabel('Movies')
plt.ylabel('Duration in min')
plt.show()
```

```python
#9.TV Show wiyh highest duration:
b=data[data['type']=='TV Show']
b['duration']=b['duration'].apply(split)
b['duration']=b['duration'].astype('int64')
b.groupby(['title'])['duration'].max().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.2f%%')
plt.title('TV Show with Highest Duration.')
plt.ylabel('Duration')
plt.show()
```

* The plot shows the TV Shows wiyh highest duration
* The duration was measured in seasons
* Supernatural TV Show has highest number of season.
  

```python
#10.average movie duration over time:
a=data[data['type']=='Movie']
a['duration']=a['duration'].apply(split)
a['duration']=a['duration'].astype('int64')
a.groupby(['release_year'])['duration'].mean().sort_values(ascending=False).head(10).plot(kind='line',marker='o',color='green')
plt.title('Average Duration of movie over time')
plt.ylabel('Duration in min')
plt.xlabel('Year');
```

* The plot shows the average duration movie based on released year.
* maximum average duration of movie is about 200min which was released over 1964

```python
#11.Count of movies added to netflix in each year:
data['added_year']=data['date_added'].dt.year
a=data[data['type']=='Movie']
a.groupby(['added_year'])['title'].count().sort_values(ascending=False).plot(kind='line',marker='o')
plt.grid()
plt.title('Total Number of movies added in year');
```

* maximum number of movies were added to netflix in the year 2019
* count of movies added is constant during 2008 to 2014

```python
#12.Average Content Duration by Director:
avg_duration=a.groupby(['director'])['duration'].mean().sort_values(ascending=False).head(10).to_frame()
sns.lineplot(x='director',y='duration',data=avg_duration,marker='o')
plt.title('Average Content Duration by Director')
plt.xticks(rotation=45);
```

* Houssam El-Din Mustafa directed movies with maximum duration

```python
#13.Rating count for genre of content:
data.groupby(['genre'])['rating'].count().sort_values(ascending=False).head(20).plot(kind='bar',color='skyblue',edgecolor='blue')
plt.title('Rating of Genre')
plt.xlabel('Genre')
plt.ylabel('Count');
```

* dramas have maximum rating whereas kids TV has minimum rating

```python
#14.correlation between released year and added year:
data['added_year']=data['date_added'].dt.year
corr=data[['release_year','added_year']].corr()
corr
```

```python
#15.correlation between release year and added year:
sns.heatmap(corr,annot=True)
plt.title('Correlation between Release Year and Added Year');
```

* there is aperfect correlation between same variable
* The value 0.0098 indicates weak correlation between the year a title was released and the year it was added to netflix
* it also indicate that the year title released does not predict when it is added to netflix.

```python
#15.Pairplot of netflix data based on type:
sns.pairplot(data,hue='type');
```

* In 1st plot shows that most of the netflix content were released after 2000 with sharp increase around 20202.
* also scatter plot indicate that most of the content released by netflix were movies
  

```python
#16.violin plot of released year:
sns.violinplot(y='release_year',data=data)
```

* this plot shows that distribution of release year is heavily concentrated from 2000 onwards

# Conclusion

* The majority of Netflix's content consists of movies, indicating a significant demand for cinematic experiences on the platform. This trend suggests that Netflix's strategy should continue to prioritize movie acquisitions and productions.
* A substantial portion of Netflix's content is produced in the United States, highlighting a potential opportunity for growth. By expanding the catalog to include more non-English content, Netflix can attract a broader global audience and tap into diverse markets.
* The prevalence of TV-MA rated content indicates that Netflix primarily caters to a mature audience. While this content appeals to a specific demographic, it may limit the platform's appeal to families and younger viewers.
* Introducing more family-friendly programming could broaden Netflix's audience base, making the platform more attractive to families and viewers seeking content suitable for all ages.
